```python
# Inicializar listas para armazenar nomes de algoritmos, acurácias e detalhes dos modelos

al = []
accuracy = []
models = []

# Função para avaliar o desempenho de um modelo

def evaluate_model(algorithm_name, model):

    # Treinar modelo
    model.fit(X_train, y_train)

    # Previsões
    y_pred = model.predict(X_test)
    y_pred_prob = model.predict_proba(X_test)

    # Accuracy
    acc_model = model.score(X_test, y_test)
    train_acc_model = model.score(X_train, y_train)

    # Métricas adicionais
    precision = precision_score(y_test, y_pred, average="weighted")

    recall = recall_score(y_test, y_pred, average="weighted")

    f1 = f1_score(y_test, y_pred, average="weighted")

    # Armazena resultados
    al.append(algorithm_name)
    accuracy.append(acc_model)

    # Print das métricas
    print('\033[31m' + f'For {algorithm_name}' + '\033[0m\n')

    print(f'Training Accuracy: {train_acc_model * 100:.4f} %\n')
    print(f'Testing Accuracy: {acc_model * 100:.4f} %\n')
    print(f'Accuracy Score: {accuracy_score(y_test, y_pred) * 100:.4f} %\n')

    print(f'Precision Score: {precision:.4f}')
    print(f'Recall Score: {recall:.4f}')
    print(f'F1 Score: {f1:.4f}\n')

    print(classification_report(y_test, y_pred))

    # Plota os gráficos

    fig, axes = plt.subplots(1, 2, figsize=(20, 8))

    fig.suptitle(
        f'Graphs for {algorithm_name}',
        fontsize=25
    )

    # Matriz de Confusão

    sns.heatmap(
        confusion_matrix(y_test, y_pred),
        annot=True,
        cmap='RdGy',
        annot_kws={'size': 15},
        square=True,
        fmt='.0f',
        ax=axes[0]
    )

    axes[0].set_title(
        'Confusion Matrix',
        fontsize=20
    )

    # Curva ROC

    lb = LabelBinarizer()
    y_test_bin = lb.fit_transform(y_test)

    for i in range(y_test_bin.shape[1]):

        fpr, tpr, _ = roc_curve(
            y_test_bin[:, i],
            y_pred_prob[:, i]
        )

        roc = auc(fpr, tpr)

        axes[1].plot(
            fpr,
            tpr,
            label=f'Class {i+1} (AUC = {roc:.2f})'
        )

    axes[1].set_xlabel(
        'False Positive Rate',
        fontsize=15
    )

    axes[1].set_ylabel(
        'True Positive Rate',
        fontsize=15
    )

    axes[1].set_title(
        'ROC Curves for Multiclass',
        fontsize=20
    )

    axes[1].legend()

    plt.tight_layout()
    plt.show()

    # Salva os resultados

    models.append(
        (
            algorithm_name,
            train_acc_model,
            acc_model,
            precision,
            recall,
            f1
        )
    )
```

### **Logistic Regression**



A regressão logística é uma técnica de classificação linear amplamente utilizada que prevê a probabilidade de ocorrência de um evento ajustando os dados a uma curva logística. Ela é particularmente adequada para tarefas de classificação binária, nas quais o resultado é uma probabilidade entre 0 e 1. Esse método funciona estabelecendo uma relação entre as características (variáveis ​​independentes) e o alvo (variável dependente), utilizando uma função sigmoide para gerar probabilidades.

### **Hiperparâmetros importantes:**

1. **Penalty (Regularização)**:

   - Este hiperparâmetro determina o tipo de regularização aplicado ao modelo. Os dois tipos principais são a regularização **L1 (Lasso)** e a **L2 (Ridge)**. A regularização L1 pode auxiliar na seleção de atributos ao levar alguns coeficientes a zero, enquanto a regularização L2 ajuda a prevenir o *overfitting* (sobreajuste) ao penalizar coeficientes elevados.



2. **C (Inverso da Intensidade da Regularização)**:

   - O parâmetro `C` controla a intensidade da regularização. Um valor baixo de `C` implica uma regularização forte, enquanto um valor mais alto significa uma regularização mais fraca. O valor ideal de `C` equilibra o bom ajuste do modelo aos dados com a prevenção do *overfitting*.



3. **Solver**:

   - O *solver* determina qual algoritmo de otimização é utilizado para minimizar a função de perda. Diferentes *solvers* são adequados para diferentes tamanhos de conjunto de dados e métodos de regularização. Exemplos comuns incluem **‘liblinear’**, **‘newton-cg’** e **‘saga’**. A escolha do *solver* pode impactar tanto o desempenho do modelo quanto a eficiência computacional.



4. **Max_iter (Número Máximo de Iterações)**:

   - Este hiperparâmetro define o número máximo de iterações que o *solver* realizará para convergir. Se o modelo não convergir antes de atingir esse limite, isso pode indicar que os dados não são linearmente separáveis ​​ou que a regularização precisa de ajustes.


```python

pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA()),
    ('model', LogisticRegression(max_iter=1000, class_weight='balanced'))
])

param_grid_lr = {
    'pca__n_components': [5, 7, 9],
    'model__C': [0.01, 0.1, 1, 10],
    'model__solver': ['liblinear', 'newton-cg', 'saga']
}

grid_lr = GridSearchCV(
    pipeline_lr,
    param_grid_lr,
    cv=5,
    scoring='balanced_accuracy',
    n_jobs=-1
)

grid_lr.fit(X_train, y_train)

print(grid_lr.best_params_)
print(grid_lr.best_score_)

evaluate_model(
    'Logistic Regression',
    grid_lr.best_estimator_
)
```

### **Random Forest**

O Random Forest é um método de aprendizado em conjunto (*ensemble learning*), o que significa que ele combina as previsões de múltiplos modelos para melhorar o desempenho geral. Nesse caso, o Random Forest utiliza diversas árvores de decisão para formar uma "floresta" de árvores. Cada árvore da floresta é treinada com um subconjunto aleatório tanto dos dados de treinamento quanto das variáveis ​​(ou *features*), o que ajuda a reduzir o *overfitting* e a aumentar a robustez do modelo. A previsão final é obtida pela agregação das previsões de todas as árvores individuais, geralmente por meio de votação (para classificação) ou cálculo de média (para regressão). Essa abordagem de conjunto contribui para a criação de um modelo mais estável e preciso.

### **Hiperparâmetros importantes:**



1. **n_estimators**: Este parâmetro controla o número de árvores na floresta. Um número maior de árvores geralmente melhora o desempenho do modelo, reduzindo a variância, mas também aumenta os custos computacionais. À medida que o número de árvores aumenta, o modelo se torna mais robusto, mas requer mais memória e poder de processamento.

2. **Critério**: O critério define como a qualidade de uma divisão é avaliada. Os dois critérios mais comuns são:

- **Impureza de Gini**: Mede a impureza de uma divisão, calculando a frequência com que uma amostra escolhida aleatoriamente seria classificada incorretamente.

- **Entropia**: Mede o ganho de informação em cada divisão, avaliando o quanto a incerteza é reduzida após a divisão.

3. **max_depth**: Este parâmetro especifica a profundidade máxima de cada árvore de decisão individual. Ao limitar a profundidade, você pode evitar o sobreajuste (overfitting), que ocorre quando uma árvore se torna muito complexa e começa a memorizar os dados de treinamento. Um valor maior para max_depth permite que as árvores capturem relações mais complexas, mas também aumenta o risco de sobreajuste.

4. **max_features**: Este parâmetro define o número de características a serem consideradas ao dividir um nó em cada árvore. A seleção aleatória de um subconjunto de características a cada divisão introduz maior diversidade entre as árvores, o que ajuda a melhorar o desempenho do modelo e a reduzir o sobreajuste. Ao ajustar o valor de max_features, você pode controlar a aleatoriedade da seleção de características, impactando tanto o desempenho quanto a velocidade do modelo.


```python
pipeline_rf = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA()),
    ('model', RandomForestClassifier(random_state=42, class_weight='balanced'))
])

param_grid_rf = {
    'pca__n_components': [5, 7, 9],
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [5, 10, 15, None]
}

grid_rf = GridSearchCV(
    pipeline_rf,
    param_grid_rf,
    cv=5,
    scoring='balanced_accuracy',
    n_jobs=-1
)

grid_rf.fit(X_train, y_train)

print(grid_rf.best_params_)
print(grid_rf.best_score_)

evaluate_model(
    'Random Forest Classifer',
    grid_rf.best_estimator_
)
```

### **Support Vector Classifier (SVC)**
O Classificador de Vetores de Suporte (SVC) é um algoritmo de aprendizado supervisionado poderoso e amplamente utilizado, projetado principalmente para tarefas de classificação. Ele visa encontrar o hiperplano ideal que melhor separa as diferentes classes no espaço de características; essa separação pode constituir uma fronteira de decisão altamente complexa, especialmente em espaços de dimensões elevadas.

Como funciona:
Em sua essência, o SVC busca encontrar um hiperplano (ou fronteira de decisão) que maximize a margem entre as diferentes classes. A margem refere-se à distância entre os pontos de dados de cada classe mais próximos do hiperplano. Ao maximizar essa margem, o SVC obtém uma melhor capacidade de generalização e minimiza erros de classificação em dados não vistos anteriormente.

### **Hiperparâmetros importantes:**


1. **C (Regularization Parameter):** The C parameter controls the trade-off between achieving a large margin and having low classification error. A smaller value of C allows for a larger margin but might allow more misclassifications, while a larger value of C penalizes misclassifications more heavily, leading to a smaller margin but fewer errors.

2. **Kernel Function**: The kernel function defines the type of decision boundary. Commonly used kernels include:

  - **Linear kernel:** Assumes a linear separation between classes.

  - **Polynomial kernel:** Suitable for non-linear relationships, where the decision boundary is a polynomial.

  - **Radial Basis Function (RBF) kernel:** Maps data into higher dimensions, allowing for highly non-linear decision boundaries.

3. **Gamma (Kernel Coefficient)**: The gamma parameter defines how much influence each training sample has on the decision boundary. A small gamma means that the influence of each data point is more spread out, leading to smoother decision boundaries. A larger gamma means that each data point has a more localized influence, leading to more complex decision boundaries. Selecting the appropriate gamma is critical for ensuring a good decision boundary.

```python
pipeline_svc = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA()),
    ('model', SVC(probability=True, class_weight='balanced'))
])

param_grid_svc = {
    'pca__n_components': [5, 7, 9],
    'model__C': [0.1, 1, 10],
    'model__kernel': ['linear', 'rbf'],
    'model__gamma': ['scale', 'auto']
}

grid_svc = GridSearchCV(
    pipeline_svc,
    param_grid_svc,
    cv=5,
    scoring='balanced_accuracy',
    n_jobs=-1
)

grid_svc.fit(X_train, y_train)

print(grid_svc.best_params_)
print(grid_svc.best_score_)

evaluate_model(
    'Support Vector Classifier',
    grid_svc.best_estimator_
)
```
